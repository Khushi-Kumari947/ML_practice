```python
# Create subplots (1 row, 3 columns)
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Box plots for each column
axes[0].boxplot(df["X"])
axes[0].set_title("X")

axes[1].boxplot(df["Y"])
axes[1].set_title("Y")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["X"], df["Y"], color="red", alpha=0.6, label="Data Points")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
X=df[['X']]
y=df['Y']
```

```python
X.ndim
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf=PolynomialFeatures(degree=2)
pf.fit(X)
X=pf.transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)*100
```

```python
prd=lr.predict(X)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["X"], df["Y"])
plt.plot(df['X'],prd,c='red')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(["org","prediction"])
plt.show()
```
